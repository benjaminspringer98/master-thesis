```python
import os
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from scipy.signal import savgol_filter
from scipy.interpolate import CubicSpline
from scipy import stats
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

#TODO Abweichung vom besten Pfad
#TODO "Glattheit" 


#Discards also first try
NUMBERRIGHT = True
MIN_MISMATCHES = 0
MAX_MISMATCHES_ALLOWED = 4
#Apply movement smoothing
SMOOTH = True

# Define the path to the "MovementFolders" directory
movement_folders_path = os.path.join(os.getcwd(), 'MovementFolders')

# Initialize a list to store the dataset records
dataset_records = []

def is_within_tolerance(eingegeben, number):
    # Convert both numbers to strings
    str_eingegeben = str(eingegeben)
    str_number = str(number)
    # Check if the lengths are the same; if not, handle accordingly
    if len(str_eingegeben) != len(str_number):
        print("ERROR: is_within_tolerance length missmatch")
        # Handle different lengths; for example, return False or handle based on your requirements
        return False  # or another appropriate handling

    # Count mismatches
    mismatches = sum(1 for a, b in zip(str_eingegeben, str_number) if a != b)
    return (mismatches <= MAX_MISMATCHES_ALLOWED) & (mismatches >= MIN_MISMATCHES)

def smooth_xyz_data(df, window_length, polyorder):
    """
    Smooths the x, y, and z data using a Savitzky-Golay filter.

    :param df: DataFrame containing the x, y, z data columns.
    :param window_length: The length of the filter window (i.e., the number of coefficients). 
                            Window length should be a positive odd integer.
    :param polyorder: The order of the polynomial used to fit the samples. 
                        Polyorder must be less than window_length.
    :return: DataFrame with smoothed x, y, and z data.
    """
    df_smoothed = df.copy()
    if SMOOTH:
        df_smoothed['x'] = savgol_filter(df['x'], window_length, polyorder)
        df_smoothed['y'] = savgol_filter(df['y'], window_length, polyorder)
        df_smoothed['z'] = savgol_filter(df['z'], window_length, polyorder)
    return df_smoothed

def read_tutorialtime_and_survey(folder_path):   
    record = {}
    # Read tutorialtime.txt
    try:
        with open(os.path.join(folder_path, 'tutorialtime.txt'), 'r') as f:
            # Read the first line and strip whitespace
            time_str = f.readline().strip()
            # Replace comma with dot if present
            time_str = time_str.replace(',', '.')
            # Convert to float
            record['tutorialtime'] = float(time_str)
    except FileNotFoundError:
        record['tutorialtime'] = None
    except ValueError:
        # Handle cases where the string cannot be converted to a float
        record['tutorialtime'] = None

    # Read survey_results.txt
    try:
        with open(os.path.join(folder_path, 'survey_results.txt'), 'r') as f:
            results = f.readlines()
            record['q1'] = results[0].strip()
            record['q2'] = results[1].strip()
            record['q3'] = results[2].strip()
            record['q4'] = results[3].strip()
            record['q5'] = results[4].strip()
            record['q6'] = results[5].strip()
            record['q7'] = results[6].strip()
            record['q8'] = results[7].strip()
            record['q9'] = results[8].strip()
    except FileNotFoundError:
        # If the file is not found, store None for each survey question
        for i in range(1, 10):
            record[f'q{i}'] = None

    return record

def read_movement_times(folder_path):    
    
    with open(os.path.join(folder_path, 'Timestamps.txt'), 'r') as file:
        timestamps_data = file.readlines()

    # Initialize lists to store parsed data
    events = []
    timestamps = []
    eingegeben_counts = []  # List to store eingegeben counts
    match_statuses =[]

    # Regular expressions to match events and timestamps
    grab_pattern = re.compile(r'GrabBuchstabenklotz\s*\((\d+)\)')
    release_pattern = re.compile(r'ReleaseBuchstabenklotz\s*\((\d+)\)')

    # Variables to keep track of the state
    grab_in_progress = None  # Store the last grab event
    eingegeben_count = 0     # Initialize the counter
    exclude_release = False  # Flag to indicate if release events should be excluded

    # Function to parse a line and extract the event name and timestamp
    def parse_line(line):
        parts = line.split(':')
        event = parts[0].strip()
        # Check if there is a timestamp and replace comma with dot if present
        timestamp = float(parts[1].strip().replace(',', '.')) if len(parts) > 1 else None
        return event, timestamp

    grabline=0
    last_number_event=0
    # Process each line in the data
    ignore=False
    for j,line in enumerate(timestamps_data):
        event, timestamp = parse_line(line)

        # Check and update the last 'Number' event
        if 'Number' in event:
            # Parse the numbers (either a single digit or a comma-separated string)
            number_part = event.split(' ', 1)[1]
            if ',' in number_part:
                last_number_event = int(''.join(number.strip() for number in number_part.split(',')))
            else:
                last_number_event = int(number_part.strip())  # Single digit, convert to integer


        # Update eingegeben_count if the event is 'eingegeben'
        if 'Eingegeben' in event:
            eingegeben_count = len(event)-11
            eingegeben_value = event.split()[1].split(':')[0]  # Extract the number part before ':'
            # Compare based on the length of the 'Number' event
            if last_number_event:
                if len(str(last_number_event)) == 1:
                    # Exact match required for single digits
                    match_status=(int(eingegeben_value) == int(last_number_event))
                elif len(str(last_number_event)) == 7:
                    # Use tolerance for seven-digit numbers
                    match_status=(is_within_tolerance(eingegeben_value, last_number_event))
                else:
                    match_status=(False)  # In case of unexpected length
            else:
                match_status=(False)  # In case there was no preceding 'Number' event                
            exclude_release = True  # Start excluding release events after 'eingegeben'
            
            for i in range(grabline) :
                    eingegeben_counts.append(eingegeben_count)  # Append the eingegeben count
                    match_statuses.append(match_status)
            ignore=True
            grabline=0
            
        
        
        # If the event is a grab event, update the grab_in_progress
        if grab_pattern.match(event):
            grab_in_progress = (event, timestamp, eingegeben_count)
        
        # If the event is a release event and we're not excluding releases
        if release_pattern.match(event):
            if grab_in_progress and not exclude_release:
                if not ignore:
                    events.append((grab_in_progress[0], event))
                    timestamps.append((grab_in_progress[1], timestamp))
                    grabline+=1
                else:
                    ignore=False
            grab_in_progress = None  # Reset after pairing with a release

        # Reset the exclude flag after release events are no longer needed to be excluded
        if exclude_release and not release_pattern.match(event):
            exclude_release = False
            # Create a DataFrame to store the results

    eingegeben_counts.pop()
    match_statuses.pop()
    events.pop()
    timestamps.pop()        

            
    df_events = pd.DataFrame(events, columns=['GrabEvent', 'ReleaseEvent'])
    df_eingegeben = pd.DataFrame(eingegeben_counts, columns=['EingegebenCount'])
    df_timestamps = pd.DataFrame(timestamps, columns=['GrabTimestamp', 'ReleaseTimestamp'])
    df_matched = pd.DataFrame(match_statuses, columns=['Matched'])

    # Combine data into a single DataFrame
    df = pd.concat([df_events, df_timestamps,df_eingegeben,df_matched], axis=1)    
    return df

def minimum_jerk_score(df):
    # Zeitpunkte und Positionen extrahieren
    times = df['time'].values
    positions = df[['x', 'y', 'z']].values

    # Zeitraum der Bewegung
    total_time = times[-1] - times[0]

    # Berechnung der Minimum Jerk Trajektorie
    # Minimum Jerk Trajektorie ist eine kubische Spline-Funktion
    time_normalized = (times - times[0]) / total_time
    start_pos = positions[0]
    end_pos = positions[-1]
    mj_trajectory = CubicSpline([0, 1], [start_pos, end_pos], bc_type=((1, [0, 0, 0]), (1, [0, 0, 0])))

    # Tatsächliche Positionen mit der Minimum Jerk Trajektorie vergleichen
    ideal_positions = mj_trajectory(time_normalized)
    diff = positions - ideal_positions
    jerk_error = np.sqrt(np.sum(diff**2, axis=1)).mean()  # Mittlerer quadratischer Fehler

    return jerk_error

def calculate_features(df_movement, grab_time, release_time):
    """
    Calculate the features of movement in a given timeframe.

    :param df_movement: DataFrame containing movement data.
    :param grab_time: Start time of the movement.
    :param release_time: End time of the movement.
    :return: Averages of the features in the given timeframe.
    """
    # Filter the dataframe for the given timeframe
    df_filtered = df_movement[(df_movement['time'] >= grab_time) & (df_movement['time'] <= release_time)]
    #smoothing Data #TODO search for nice filter values
    df_smooth=smooth_xyz_data(df_filtered,7,3)
    # Calculate differences in position and time
    df_diff = df_smooth.diff().iloc[1:]  # Exclude the first row which is NaN after diff()
    df_diff['dt'] = df_diff['time']  # Time difference
    #TODO gleiche Zeiten rausfiltern?
    df_diff = df_diff[df_diff['dt'] >= 0.012]

    df_diff['dx'] = np.sqrt(df_diff['x']**2 + df_diff['y']**2 + df_diff['z']**2)  # Distance difference
    # Calculate speed (distance/time)
    df_diff['speed'] = df_diff['dx'] / df_diff['dt']

    # Calculate jerk_error
    average_jerk_error = minimum_jerk_score(df_smooth)
    


    total_distance = df_diff['dx'].sum()
    if (average_jerk_error > 500)&(total_distance>0.1):
        print("jerk_error: "+str(average_jerk_error))
        print("total_distance: "+ str(total_distance))
    
    total_time=release_time-grab_time
    # Return the average speed
    return total_distance, total_time, average_jerk_error

def get_values(folder_path, movement_timestamps):
    """
    The function `get_values` extracts movement data from specified files, computes features such as
    distance, speed, and jerk_error for each movement interval, and returns a DataFrame with the computed
    features.
    
    :param folder_path: The `folder_path` parameter is the path to the folder containing the movement
    data files. It should be a string representing the directory path where the data files are located
    :param movement_timestamps: The `movement_timestamps` parameter is a DataFrame that contains the
    timestamps for each movement interval. It has two columns: 'GrabTimestamp' and 'ReleaseTimestamp'.
    Each row represents a movement interval, with the 'GrabTimestamp' indicating the start time of the
    interval and the 'ReleaseTimestamp' indicating
    :return: a DataFrame with computed features for each movement interval. The features include
    distances, speeds, and jerk_errors for the right hand, head, and left hand.
    """
    
    right_hand_file = os.path.join(folder_path,'data_RightHand.txt')
    head_file = os.path.join(folder_path,'data_Head.txt')
    left_hand_file = os.path.join(folder_path,'data_LeftHand.txt')
    
    # Column names for the movement data
    movement_columns = ["time", "x", "y", "z", "rotationx", "rotationy", "rotationz", "rotationw"]
    def convert_decimal(column):
        return column.apply(lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)


    # Reading the files into dataframes
    df_right_hand = pd.read_csv(right_hand_file, sep="\s+", header=None, names=movement_columns)
    df_head = pd.read_csv(head_file, sep="\s+", header=None, names=movement_columns)
    df_left_hand = pd.read_csv(left_hand_file, sep="\s+", header=None, names=movement_columns)
    
    decimal_columns = ["time","x", "y", "z", "rotationx", "rotationy", "rotationz", "rotationw"]
    for column in decimal_columns:
        df_right_hand[column] = convert_decimal(df_right_hand[column])
        df_head[column] = convert_decimal(df_head[column])
        df_left_hand[column] = convert_decimal(df_left_hand[column])

    # Calculate average speeds for each interval for right hand, head, and left hand if long enough
    MIN_DURATION = 0.1
    average_values = []
    for index, row in movement_timestamps.iterrows():
        grab_time = row['GrabTimestamp']
        release_time = row['ReleaseTimestamp']
        duration = release_time - grab_time
        if duration >= MIN_DURATION:
            distance_right,time_right, average_jerk_error_right = calculate_features(df_right_hand, grab_time, release_time)
            distance_head, time_head, average_jerk_error_head = calculate_features(df_head, grab_time, release_time)
            distance_left, time_left, average_jerk_error_left = calculate_features(df_left_hand, grab_time, release_time)


            if distance_left > distance_right:
                distance_merker = distance_right
                distance_right = distance_left
                distance_left = distance_merker
                average_jerk_error_merker = average_jerk_error_right
                average_jerk_error_right = average_jerk_error_left
                average_jerk_error_left = average_jerk_error_merker
            speed_right_hand = distance_right/time_right
            speed_head = distance_head/time_head
            speed_left_hand = distance_left/time_left
            
            average_values.append({
                'Interval': index,
                'Distance_RightHand': distance_right,
                'Distance_Head': distance_head,
                'Distance_LeftHand': distance_left,
                'Speed_RightHand': speed_right_hand,
                'Speed_Head': speed_head,
                'Speed_LeftHand': speed_left_hand,
                'Curve_RightHand': average_jerk_error_right,
                'Curve_Head': average_jerk_error_head,
                'Curve_LeftHand': average_jerk_error_left,
                'time':time_head
            })

    average_values_df = pd.DataFrame(average_values)
    average_values_df["CNT"]=movement_timestamps["EingegebenCount"]
    average_values_df["MATCH"] = movement_timestamps["Matched"]

    return average_values_df

def comparative_stats(df1, df7, label, tracked):
    print("\n")
    print(tracked)
    print(label)
    print("Nur 1: \t" + str(df1.mean()))
    print("7: \t" + str(df7.mean()))
    print("Diff: \t" + str((df7-df1).mean()))
    print("\n")
    mean1, std1 = df1.mean(), df1.std()
    mean2, std2 = df7.mean(), df7.std()
    
    # Perform a t-test
    df1 = df1.dropna()
    df7 = df7.dropna()

    # Perform t-test on cleaned data
    t_stat, p_value = stats.ttest_ind(df1, df7, equal_var=False)

    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")
    # Interpretation of the t-test result
    alpha = 0.05
    if p_value < alpha:
        print("🎉🥳🎈🎊🍾There is a significant difference between the groups.🎉🥳🎈🎊🍾")
    else:
        print("There is no significant difference between the groups.")

    # Punkte für die Gauß-Kurven generieren
    x1 = np.linspace(min(df1), max(df1), 100)
    y1 = (1 / (np.sqrt(2 * np.pi) * std1)) * np.exp(-0.5 * ((x1 - mean1) / std1)**2)

    x2 = np.linspace(min(df7), max(df7), 100)
    y2 = (1 / (np.sqrt(2 * np.pi) * std2)) * np.exp(-0.5 * ((x2 - mean2) / std2)**2)

    # Erster Plot (Histogramm mit Gauß-Kurven)
    plt.figure(figsize=(12, 5))  # Gesamtgröße des Plots

    # Erster Subplot
    plt.subplot(1, 2, 1)  # (Reihen, Spalten, aktueller Plot)
    plt.hist(df1, bins=10, alpha=0.5, label=label + "_" + tracked +"_1", color='blue')
    plt.hist(df7, bins=10, alpha=0.5, label=label + "_" + tracked +"_7", color='green')
    plt.plot(x1, y1, color='darkblue', linewidth=2)
    plt.plot(x2, y2, color='darkgreen', linewidth=2)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {label}_{tracked} with Gaussian Curves')
    plt.legend()

    # Berechnung der Differenz
    max_val = max(df1.max(), df7.max())
    diff = (df1-df7)/max_val
    mean, std = diff.mean(), diff.std()

    # Punkte für die Gauß-Kurve der Differenz generieren
    x = np.linspace(min(diff), max(diff), 100)
    y = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-0.5 * ((x - mean) / std)**2)

    # Zweiter Subplot
    plt.subplot(1, 2, 2)  # (Reihen, Spalten, aktueller Plot)
    plt.hist(diff, bins=10, alpha=0.5, label=label + "_" + tracked +"_Diff", color='red')
    plt.plot(x, y, color='darkred', linewidth=2)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Diff {label}_{tracked} with Gaussian Curves')
    plt.legend()

    # Gesamten Plot anzeigen
    plt.tight_layout()
    plt.show()
    


# Iterate through each folder in the MovementFolders directory
for folder_name in os.listdir(movement_folders_path):
    # Skip if it's not a directory or if it's a hidden/system directory (starts with '.')
    if not os.path.isdir(os.path.join(movement_folders_path, folder_name)) or folder_name.startswith('.'):
        continue

    folder_path = os.path.join(movement_folders_path, folder_name)
    print(folder_name)
    # Read the tutorialtime and survey_results files
    record = read_tutorialtime_and_survey(folder_path)
    record['folder_name'] = folder_name  # Store the folder name in the record

    movement_timestamps= read_movement_times(folder_path) 
    values = get_values(folder_path, movement_timestamps)
    
    conditions = (values['Speed_RightHand'] < 0.01) | (values['Speed_LeftHand'] < 0.001) | (values['time'] < 0.1) | (values['Distance_RightHand'] < 0.1) | (values['Distance_LeftHand'] < 0.001)
    values = values[~conditions]
    #df_all_movements=df_all_movements.append(values)
    if NUMBERRIGHT:
        condition2 = (values["MATCH"]==False)
        values = values[~condition2]
    
    if ("Speed_RightHand" in values.columns) & ("Speed_LeftHand" in values.columns):    
        record["speed_right_1"] = values[values["CNT"]==1]["Speed_RightHand"].mean()
        record["speed_right_7"] = values[values["CNT"]==7]["Speed_RightHand"].mean()
        record["speed_left_1"] = values[values["CNT"]==1]["Speed_LeftHand"].mean()
        record["speed_left_7"] = values[values["CNT"]==7]["Speed_LeftHand"].mean()
        record["speed_head_1"] = values[values["CNT"]==1]["Speed_Head"].mean()
        record["speed_head_7"] = values[values["CNT"]==7]["Speed_Head"].mean()

        record["distance_right_1"] = values[values["CNT"] == 1]["Distance_RightHand"].mean()
        record["distance_right_7"] = values[values["CNT"] == 7]["Distance_RightHand"].mean()
        record["distance_left_1"] = values[values["CNT"] == 1]["Distance_LeftHand"].mean()
        record["distance_left_7"] = values[values["CNT"] == 7]["Distance_LeftHand"].mean()
        record["distance_head_1"] = values[values["CNT"] == 1]["Distance_Head"].mean()
        record["distance_head_7"] = values[values["CNT"] == 7]["Distance_Head"].mean()

        record["average_jerk_error_right_1"] = values[values["CNT"] == 1]["Curve_RightHand"].mean()
        record["average_jerk_error_right_7"] = values[values["CNT"] == 7]["Curve_RightHand"].mean()
        record["average_jerk_error_left_1"] = values[values["CNT"] == 1]["Curve_LeftHand"].mean()
        record["average_jerk_error_left_7"] = values[values["CNT"] == 7]["Curve_LeftHand"].mean()
        record["average_jerk_error_head_1"] = values[values["CNT"] == 1]["Curve_Head"].mean()
        record["average_jerk_error_head_7"] = values[values["CNT"] == 7]["Curve_Head"].mean()
    else:
        print("missing either right hand speed ot left hand speed")
    """
    
    
    if ("Speed_RightHand" in values.columns) & ("Speed_LeftHand" in values.columns):    
        record["speed_right_1"] = values[values["CNT"]==1]["Speed_RightHand"].median()
        record["speed_right_7"] = values[values["CNT"]==7]["Speed_RightHand"].median()
        record["speed_left_1"] = values[values["CNT"]==1]["Speed_LeftHand"].median()
        record["speed_left_7"] = values[values["CNT"]==7]["Speed_LeftHand"].median()
        record["speed_head_1"] = values[values["CNT"]==1]["Speed_Head"].median()
        record["speed_head_7"] = values[values["CNT"]==7]["Speed_Head"].median()

        record["distance_right_1"] = values[values["CNT"] == 1]["Distance_RightHand"].median()
        record["distance_right_7"] = values[values["CNT"] == 7]["Distance_RightHand"].median()
        record["distance_left_1"] = values[values["CNT"] == 1]["Distance_LeftHand"].median()
        record["distance_left_7"] = values[values["CNT"] == 7]["Distance_LeftHand"].median()
        record["distance_head_1"] = values[values["CNT"] == 1]["Distance_Head"].median()
        record["distance_head_7"] = values[values["CNT"] == 7]["Distance_Head"].median()

        record["average_jerk_error_right_1"] = values[values["CNT"] == 1]["Curve_RightHand"].median()
        record["average_jerk_error_right_7"] = values[values["CNT"] == 7]["Curve_RightHand"].median()
        record["average_jerk_error_left_1"] = values[values["CNT"] == 1]["Curve_LeftHand"].median()
        record["average_jerk_error_left_7"] = values[values["CNT"] == 7]["Curve_LeftHand"].median()
        record["average_jerk_error_head_1"] = values[values["CNT"] == 1]["Curve_Head"].median()
        record["average_jerk_error_head_7"] = values[values["CNT"] == 7]["Curve_Head"].median()
    else:
        print("missing either right hand speed ot left hand speed")
    """





    # Add the record to the dataset records list
    dataset_records.append(record)



# Create a DataFrame from the records
dataset_df = pd.DataFrame(dataset_records)
dataset_df.dropna()
dataset_df.to_csv("data.csv", sep=";", decimal=",")

# Print the DataFrame to verify its contents
print(dataset_df)
comparative_stats(dataset_df["speed_right_1"], dataset_df["speed_right_7"], "speed", "rechts")
comparative_stats(dataset_df["speed_left_1"], dataset_df["speed_left_7"], "speed", "links")
comparative_stats(dataset_df["speed_head_1"], dataset_df["speed_head_7"], "speed", "Kopf")

comparative_stats(dataset_df["distance_right_1"], dataset_df["distance_right_7"], "Distanz", "rechts")
comparative_stats(dataset_df["distance_left_1"], dataset_df["distance_left_7"], "Distanz", "links")
comparative_stats(dataset_df["distance_head_1"], dataset_df["distance_head_7"], "Distanz", "Kopf")

comparative_stats(dataset_df["average_jerk_error_right_1"], dataset_df["average_jerk_error_right_7"], "jerk_error", "rechts")
comparative_stats(dataset_df["average_jerk_error_left_1"], dataset_df["average_jerk_error_left_7"], "jerk_error", "links")
comparative_stats(dataset_df["average_jerk_error_head_1"], dataset_df["average_jerk_error_head_7"], "jerk_error", "Kopf")
```
